### Second project
### Covid positive X-Ray image data
### https://github.com/realdfy/QMSS-5074-Portfolio/blob/main/HW2/5074_HW_2.ipynb

This project is a classication project with image data. It has three categories which are covid positivity, normal, and viral pneumonia. Building this predictive model can be used for early detection and diagnosis of covid-19.

In this project I tried CNN, transfer learning from pretrained VGG16 model and data augmentation model. They all performed well, with data augmentation outperforming the rest.

```python
covid = pd.read_csv('covid-xray.csv')

covid.head()
```

```python
import matplotlib.pyplot as plt
covid['TRANSFER_LEARNING'] = covid['TRANSFER_LEARNING'].fillna("False")
covid.boxplot(column='F1_SCORE', by=['TRANSFER_LEARNING'])
plt.title('Distribution of Scores by Transfer Learning')
plt.xlabel('Transfer Learning')
plt.ylabel('F1 Score')
plt.show()
```

By comparing to the work of others on related competition leaderboards, we can see that overall, models that use transfer learning outperform those that do not.

### Third project

This is a text classification project. The data is called Stanford Sentiment Treebank (SST), which is used for sentiment analysis, specifically in the field of movie reviews. This dataset contains two labels which are "positive" and "negative".

In this project, first we use preprocessor function to tokenize words and make sure all documents have the same length, here is 40 words. Then I implemented Embedding layers, LSTM layers, Conv1d layers, and transfer learning with glove embeddings to predict the SST sentiment dataset.

```python
sst = pd.read_csv('sstsentiment.csv')

sst.head()
```

```python
import seaborn as sns

sns.boxplot(x='lstm_layers', y='f1_score', data=sst)
```

By comparing to the work of others on related competition leaderboards, we can see that one LSTM layer performed the best on average.
